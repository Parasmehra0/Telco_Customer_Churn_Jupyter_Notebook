```python
ax = sns.countplot(x = 'SeniorCitizen', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer by Senior Citizen")
plt.show()
```

```python


# Create the count plot
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='SeniorCitizen', data=df, hue='Churn')

# Get total counts for percentage calculation
total = len(df)

# Loop through containers and add % labels
for container in ax.containers:
    labels = []
    for bar in container:
        height = bar.get_height()
        percent = (height / total) * 100
        labels.append(f'{percent:.1f}%')  # One decimal place
    ax.bar_label(container, labels=labels)

plt.title('Churn By Senior Citizen (%)')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(20, 6))
# Step 1: Create a crosstab (counts)
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Step 2: Convert counts to percentages
ct_percent = ct.div(ct.sum(axis=1), axis=0) * 100

# Step 3: Plot as stacked bar chart
ax = ct_percent.plot(kind='bar', stacked=True, colormap='Set2', figsize=(6, 4))

# Add percentage labels on each segment
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')

plt.title('Churn % by Senior Citizen')
plt.ylabel('Percentage')
plt.xlabel('Senior Citizen')
plt.legend(title='Churn', loc='upper right')
plt.tight_layout()
plt.xticks(rotation=0)
plt.show()
```

***# A greater percentage of people in senior citizen have churned.***

```python
# List of categorical columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

import matplotlib.pyplot as plt
import seaborn as sns

# List of categorical columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the subplot grid
n_cols = 3  # 3 plots per row
n_rows = -(-len(cols) // n_cols)  # Ceiling division

plt.figure(figsize=(18, 15))  # Adjust size as needed

for idx, col in enumerate(cols, 1):
    plt.subplot(n_rows, n_cols, idx)
    ax = sns.countplot(data=df, x=col, hue=col, palette="Set2", legend=False)
    
    # Label bars inside the loop
    for container in ax.containers:
        ax.bar_label(container)
    
    plt.title(f'{col} Count')
    plt.xticks(rotation=15)
    plt.tight_layout()

plt.show()
```
